```python
import numpy as np
```

```python
X=np.array([10,20,30,40,50])
Y=np.array([50,70,90,110,1030])
```

```python
def gradient_desc(X,Y):  #Define a function
  a1_curr=a0_curr=0      #Initial value of a1_curr and a0_curr
  iterations=1000      #Initialize number of steps
  n=len(X)             #Number of data points n
  learning_rate=0.1    # Initialize learning rate

  for j in range (iterations):
    y_predicted=a1_curr*X+a0_curr  # equation of line (a1X+a0)
    cost=(1/n)*sum([val**2 for val in (Y-y_predicted)]) # cost function equation
    a1_d=-(2/n)*sum(X*(Y-y_predicted))  # Partial derivative wrt slope
    a0_d=-(2/n)*sum(Y-y_predicted)   #Partial derivative wrt sintercept
    a1_curr=a1_curr-learning_rate*a1_d
    a0_curr=a0_curr-learning_rate*a0_d
    print("a1 {},a0 {}, cost{},iteration {}".format(a1_curr,a0_curr,cost,j))




```

```python
gradient_desc(X,Y)
```
