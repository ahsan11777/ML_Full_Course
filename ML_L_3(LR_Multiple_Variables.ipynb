```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

## 1.Data preprocsiing(Handling NA values)

```python
import math
median_bedrooms=math.floor(df.bedrooms.median())
median_bedrooms


```

## math.floor(This function from the Python math module takes a number as input and returns the largest integer less than or equal to that number.)

```python
df.bedrooms=df.bedrooms.fillna(median_bedrooms)
df.bedrooms
```

```python
df
```

# 2.Linear Regression using Multiple variables


```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Given these home prices find out price of a home that has
## 1.3000 sqr fit area, 3 bedrooms , 40 years old
## 2. 2500 sqr fit area, 4 bedrooms , 5 years old

#1.


```python
reg.coef_
```

```python
reg.predict([[3000,3,40]])
```

#2.


```python
reg.predict([[2500,4,5]])
```
